## IMDB Îç∞Ïù¥ÌÑ∞

### Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨

```
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.layers import Dense, Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D
```

```
num_words = 10000
max_len = 500
batch_size = 32

(input_train, y_train), (input_test, y_test) = imdb.load_data(num_words = num_words)

print(len(input_train))
print(len(input_test))
```

*imdb.data_load()Ïùò Ïù∏ÏûêÎ°ú num_wordsÎ•º ÏÇ¨Ïö©ÌïòÎ©¥ Ïù¥ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Îì±Ïû• ÎπàÎèÑ ÏàúÏúÑÎ°ú Î™á Îì±ÍπåÏßÄÏùò Îã®Ïñ¥Î•º ÏÇ¨Ïö©Ìï† Í≤ÉÏù∏ÏßÄÎ•º ÏùòÎØ∏Ìï©ÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥ÏÑú 10,000ÏùÑ ÎÑ£ÏúºÎ©¥, Îì±Ïû• ÎπàÎèÑ ÏàúÏúÑÍ∞Ä 1~10,000Ïóê Ìï¥ÎãπÌïòÎäî Îã®Ïñ¥Îßå ÏÇ¨Ïö©ÌïòÍ≤å Îê©ÎãàÎã§. Ï¶â, Îã®Ïñ¥ ÏßëÌï©Ïùò ÌÅ¨Í∏∞Îäî 10,000Ïù¥ Îê©ÎãàÎã§.*

```
pad_x_train = pad_sequences(input_train, maxlen = max_len)
pad_x_test = pad_sequences(input_test, maxlen = max_len)

print(pad_x_train)
print(pad_x_test)
```

### Î™®Îç∏ Íµ¨ÏÑ±

```
def build_model():
    model = Sequential()

    model.add(Embedding(input_dim = num_words, output_dim = 32, input_length = max_len)) # input_dim Ïú†ÎãàÌÅ¨Ìïú Îã®Ïñ¥Í∞úÏàò
    model.add(Conv1D(32, 7, activation = 'relu'))
    model.add(MaxPooling1D(7))
    model.add(Conv1D(32, 5, activation = 'relu'))
    model.add(MaxPooling1D(5))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(1, activation = 'sigmoid')) # Í∏çÎ∂ÄÏ†ï

    model.compile(optimizer = RMSprop(learning_rate = 1e-4),
                  loss = 'binary_crossentropy',
                  metrics = ['accuracy'])
    
    return model
```

```
model = build_model()
model.summary()
```

### Î™®Îç∏ ÌïôÏäµ

```
history = model.fit(pad_x_train, y_train,
                    batch_size = 128, epochs = 30,
                    validation_split = 0.2)
```

### ÏãúÍ∞ÅÌôî

```
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

loss = history.history['loss']
val_loss = history.history['val_loss']
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'b--', label = 'training loss')
plt.plot(epochs, val_loss, 'r:', label = 'validation loss')
plt.grid()
plt.legend()

plt.figure()
plt.plot(epochs, acc, 'b--', label = 'training accuracy')
plt.plot(epochs, val_acc, 'r:', label = 'validation accuracy')
plt.grid()
plt.legend()

plt.show()
```

```
model.evaluate(pad_x_test, y_test)
```

```
from tensorflow.keras.datasets import reuters
```

```
num_words = 10000
max_len = 500

(input_train, y_train), (input_test, y_test) = reuters.load_data(num_words = num_words)

print(len(input_train))
print(len(input_test))
```

*num_wordsÎäî Ïù¥ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Îì±Ïû• ÎπàÎèÑ ÏàúÏúÑÎ°ú Î™á Î≤àÏß∏Ïóê Ìï¥ÎãπÌïòÎäî Îã®Ïñ¥ÍπåÏßÄÎßå ÏÇ¨Ïö©Ìï† Í≤ÉÏù∏ÏßÄ Ï°∞Ï†àÌï©ÎãàÎã§.
```
pad_x_train = pad_sequences(input_train, maxlen = max_len)
pad_x_test = pad_sequences(input_test, maxlen = max_len)

print(pad_x_train)
print(pad_x_test)
```

### Î™®Îç∏ Íµ¨ÏÑ±

```
def build_model():
    model = Sequential()

    model.add(Embedding(input_dim = num_words, output_dim = 64, input_length = max_len))
    model.add(Conv1D(64, 7, activation = 'relu'))
    model.add(MaxPooling1D(7))
    model.add(Conv1D(64, 5, activation = 'relu'))
    model.add(MaxPooling1D(5))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(46, activation = 'softmax')) # reutersÎç∞Ïù¥ÌÑ∞Îäî Îã§Ï§ëÎ∂ÑÎ•ò

    model.compile(optimizer = 'adam',
                  loss = 'sparse_categorical_crossentropy',
                  metrics = ['accuracy'])
    
    return model
```

```
model = build_model()
model.summary()
```

### Î™®Îç∏ ÌïôÏäµ

```
history = model.fit(pad_x_train, y_train,
                    batch_size = 128, epochs = 30,
                    validation_split = 0.2)
```

### ÏãúÍ∞ÅÌôî

```
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

loss = history.history['loss']
val_loss = history.history['val_loss']
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'b--', label = 'training loss')
plt.plot(epochs, val_loss, 'r:', label = 'validation loss')
plt.grid()
plt.legend()

plt.figure()
plt.plot(epochs, acc, 'b--', label = 'training accuracy')
plt.plot(epochs, val_acc, 'r:', label = 'validation accuracy')
plt.grid()
plt.legend()

plt.show()
```

```
model.evaluate(pad_x_test, y_test)
```
